# Data Cleaning

```python
#import dependencies
import pandas as pd
```

```python
#read original data and get impressions of structure
data = pd.read_csv("movies.csv")
data.head()
```

```python
#drop columns not needed and Na
data.drop(columns=['id','imdb_votes', 'tmdb_popularity', 'imdb_id'], axis=1, inplace=True)
data.dropna(subset=['imdb_score','tmdb_score'],inplace=True)
data
```

```python
#get a column with the average score and drop the other two scores; and dropping shows with no runtime
data['Score'] = (data['imdb_score'] + data['tmdb_score'])/2
data = data.drop(columns=['imdb_score', 'tmdb_score'])
data = data.drop(data[data['runtime'] == 0].index)
data
```

```python
#data correction
data['genres'] = data['genres'].str.strip("[]'")
data['genres'] = data['genres'].str.replace("', '", ",")
data['production_countries'] = data['production_countries'].str.strip("[]'")
data['production_countries'] = data['production_countries'].str.replace("', '", ",")
data
```

```python
data['description'] = data['description'].fillna('Not description provided')
data['age_certification'] = data['age_certification'].fillna('Not Rated')
data['production_countries'] = data['production_countries'].replace("", "Not informed")
data['seasons'] = data['seasons'].fillna(0)
data['genres'] = data['genres'].fillna('Not genres')
data
```

```python
#check for NaN values
data.isna().value_counts()
```

```python
data.dropna()
```

```python
data['genres'] = data['genres'].replace("", "comedy")
```

```python
#putting data types
data['seasons'] = data['seasons'].astype(int)
data['production_countries'] = data['production_countries'].astype(object)
data['genres'] = data['genres'].astype(object)
data.dtypes
```

```python
#save data to a csv
data.to_csv("clean_movies.csv",index=False)
```
